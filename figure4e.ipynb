```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def process_xvg_file(filename):
    data = []
    
    with open(filename,"r") as file: #opens the file
        for line in file:
            if line.startswith(("#","@")): #ignores line starting with # and @
                continue
            values = line.split()
            try:
                numbers = [value for value in values]
                data.append(numbers)
            except ValueError:
                pass
    return np.array(data)
```
