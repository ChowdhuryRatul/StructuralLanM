```python
def sum_calc(weights, kd_file, data_4):
    list_ = []
    for i in range(data_4.shape[0]):
        list_.append(np.sum(data_4[i]*weights))

    list_ = np.array(list_/min(list_))

    file = []
    for i in list_:
        rank = []
        for j in list_:
            rank.append((i-j))
        file.append(rank)

    # pd.DataFrame(abs(np.array(file) - np.array(kd_file)))
    np_kd = np.array(kd_file)/np.max(np.array(kd_file))
    np_eW = np.array(file)/np.max(np.array(file))
    return np.sum(abs(np_kd-np_eW))
```

```python
sum_calc(weights, kd_file, data_4)
```

```python
frac = np.linspace(.1,1,10)
mult = np.linspace(2,10,9)
frac_mult = np.concatenate((frac,mult), axis = 0)
frac_mult
```

```python
dict_ = {}
counter = 0
for i in frac_mult:
    for j in frac_mult:
        for k in frac_mult:
            for l in frac_mult:
                inside_dict = {}
                inside_dict['w1'] = i
                inside_dict['w2'] = j
                inside_dict['w3'] = k
                inside_dict['w4'] = l 
                
                sum_ = sum_calc(np.array([i,j,k,l]), kd_file, data_4)
                inside_dict['sum'] = sum_
                dict_[counter] = inside_dict
                counter+=1
```

```python
main_data = pd.DataFrame(dict_).transpose().sort_values(by = 'sum')
```
